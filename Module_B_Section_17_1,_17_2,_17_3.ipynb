```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
import numpy as np

plt.style.use('seaborn-poster')
```

**Section 17.1(Interpolation Problem Statement)**

An interpolation function is an estimation function used to estimate a new y when given a new x, in a data set consisting of independent data values, x(i), and dependent data values, y(i), where i=1,â€¦,n.

Interpolation doesn't use an underlying model for the data, especially when there are many reliable points.

```python
x = [0, 1, 2, 3]
y = [1, 3, 5, 2]

plt.figure(figsize = (10,8))
plt.plot(x, y, 'ob')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**Section 17.2(Linear Interpolation)**

The estimated point in linear interpolation is assumed to lie on the line joining the nearest point to the left and right.

The linear interpolation at x is:
Screenshot 2024-01-20 193327.png
Assuming that the x-data points are in ascending order and that x is a point such that x(i) < x < x(i+1)

```python
#Finding the linear interpolation at x = 2.5 based on the data x = [0, 1, 2, 3], y = [1, 3, 5, 2].
x = [0, 1, 2, 3]
y = [1, 3, 5, 2]

f = interp1d(x, y)
y_hat = f(2.5)
print(y_hat)

plt.figure(figsize = (10,8))
plt.plot(x, y, '-ob')
plt.plot(2.5, y_hat, 'ro')
plt.title('Linear Interpolation at x = 2.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**Section 17.2(Cubic Spline Interpolation)**

The interpolating function in cubic spline interpolation is a set of piecewise cubic functions. Which means we assume that the points (x(i),y(i))and (x(i+1),y(i+1)) are joined by a cubic polynomial Si(x)=a(i)x^3+b(i)x^2+c(i)x+d(i), which is valid for x(i) â‰¤ x â‰¤ x(i+1) for i=1,â€¦,nâˆ’1.

Determining the coefficients a(i),b(i),c(i),d(i) for each of the cubic functions is the first step to determine the interpolating fucntion.

```python
#Plotting the cubic spline interpolation of the dataset x = [0, 1, 2, 3], y = [1, 3, 5, 2] for 0 â‰¤ x â‰¤ 3
x = [0, 1, 2, 3]
y = [1, 3, 5, 2]

f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 3, 100)
y_new = f(x_new)

plt.figure(figsize = (10,8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
