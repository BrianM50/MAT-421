```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

**Section 20.1(Numerical Differentiation Problem Statement)**

An evenly spaced set of points over the domain of a function, over some interval is called a numerical grid. The distance between adjacent points on the numerical grid is called a step size.

A function f(x) in python can be represented to be discrete, which is useful for numerical calculations.

Our goal is to determine methods of approximating the derivative of f(x) over a numerical grid and determine its accuracy.

**Section 20.2(Finite Difference Approximating Derivatives)**

The derivative f'(x) of a function f(x) at the point x = a can be found using the formula:

f'(a) = limxâ†’a (f(x)-f(a))/(x-a)

In finite difference approximations of this slope, values of the function in the neighborhood of point x=a can be used to find the derivative at x=a. There are many finite difference formulas, three of these are where the derivative is calculated using values of two points as shown below.

Forward difference: f'(x(j)) = (f(x(j+1))-f(x(j)))/(x(j+1)-x(j))

Backward difference: f'(x(j)) = (f(x(j))-f(x(j-1)))/(x(j)-x(j-j))

Central difference: f'(x(j)) = (f(x(j+1))-f(x(j-1)))/(x(j+1)-x(j-1))

```python
#Compute the derivative of sin(x) numerically

# step size
h = 0.1
# define grid
x = np.arange(0, 2*np.pi, h)
# compute function
y = np.sin(x)

# compute vector of forward differences
forward_diff = np.diff(y)/h
# compute corresponding grid
x_diff = x[:-1:]
# compute exact solution
exact_solution = np.cos(x_diff)

# Plot solution
plt.figure(figsize = (12, 8))
plt.plot(x_diff, forward_diff, '--', \
         label = 'Finite difference approximation')
plt.plot(x_diff, exact_solution, \
         label = 'Exact solution')
plt.legend()
plt.show()

# Compute max error between
# numerical derivative and exact solution
max_error = max(abs(exact_solution - forward_diff))
print('Max error for step size 0.1 =', max_error)
```

```python
#Smaller step size
h2=0.001

# define grid
x2 = np.arange(0, 2*np.pi, h2)
# compute function
y2 = np.sin(x2)

# compute vector of forward differences
forward_diff2 = np.diff(y2)/h2
# compute corresponding grid
x_diff2 = x2[:-1:]
# compute exact solution
exact_solution2 = np.cos(x_diff2)

# Plot solution
plt.figure(figsize = (12, 8))
plt.plot(x_diff2, forward_diff2, '--', \
         label = 'Finite difference approximation')
plt.plot(x_diff2, exact_solution2, \
         label = 'Exact solution')
plt.legend()
plt.show()

# Compute max error between
# numerical derivative and exact solution
max_error2 = max(abs(exact_solution2 - forward_diff2))
print('Max error for step size 0.001 =',max_error2)
```

**Section 20.4(Numerical Differentiation with Noise)**

Data can sometimes contain noise, which means its value can be shifted a small amount than from computing a pure mathemathical function.

To better understand this we can compute the derivative of two functions a regular sin function f(x) = sin(x) and a sin function corrupted by a cos function, f(x) = sin(x) + Ïµcos(Ï‰x)

```python
#Plot the two functions above when Ïµ = 0.01 and Ï‰ = 100
x = np.arange(0, 2*np.pi, 0.01)
# compute function
omega = 50
epsilon = 0.01

y = np.sin(x)
y_noise = y + epsilon*np.cos(omega*x)

# Plot solution
plt.figure(figsize = (12, 8))
plt.plot(x, y_noise, 'r-', \
         label = 'sin(x) + noise')
plt.plot(x, y, 'b-', \
         label = 'sin(x)')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.show()
```

The two functions above seem very similar but since ÏµÏ‰ may not be small when Ï‰ is large, the effect of the noise on the derivative may be large. This means that the derivative of the function may not be usable.

```python
x = np.arange(0, 2*np.pi, 0.01)
# compute function
y = np.cos(x)
y_noise = y - epsilon*omega*np.sin(omega*x)

# Plot solution
plt.figure(figsize = (12, 8))
plt.plot(x, y_noise, 'r-', \
         label = 'Derivative sin(x) + noise')
plt.plot(x, y, 'b-', \
         label = 'Derivative of sin(x)')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.show()
```
