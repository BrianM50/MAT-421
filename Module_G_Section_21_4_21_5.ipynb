**Section 21.4(Simpson's Rule)**

Simpson's rule approximates the area under the function f(x) over the subintervals, [x(i-1), x(i)] and [x(i), x(i+1)] by fitting a ploynomial through the points (x(i-1), f(x(i-1))), (x(i), f(x(i))), and (x(i+1), f(x(i+1))).

```python
import numpy as np

#Approximate ‚à´[2œÄ,0] cos(x)dx with 15 evenly spaced grid points over the whole interval, using Simpson's Rule.

a = 0
b = 2*np.pi
n = 15
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.cos(x)

I_simp = (h/3) * (f[0] + 2*sum(f[:n-2:2]) + 4*sum(f[1:n-1:2]) + f[n-1])
err_simp = 0 - I_simp

print('Simpson\'s Rule:', I_simp)
print('Error:', err_simp)
```

**Section 21.5(Computing Integrals in Python)**

A sub-package that has several functions for computing integrals is scipy.integrate.

One of these functions is called trapz, it takes an array of function values f computed on a numerical grid x as input arguments.

```python
#Approximate ‚à´[2œÄ,0] cos(x)dx with 15 evenly spaced grid points over the whole interval, using the trapz function.

from scipy.integrate import trapz

a = 0
b = 2*np.pi
n = 15
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.cos(x)

I_trapz = trapz(f,x)
I_trap = (h/2)*(f[0] + 2 * sum(f[1:n-1]) + f[n-1])

print('Trapz function:', I_trapz)
print('Trapezoid Rule:', I_trap)
```

To find the approximated cumulative integral, use the cumtrapz function cumsum. This function takes the same input arguments as the function trapz.

```python
#Approximate the cumulative integral of ùëì(ùë•)=cos(y) from 0 to 2ùúã with a discretization step of 0.1, using the cumtrapz function.

from scipy.integrate import cumtrapz
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-poster')

x = np.arange(0, 2*np.pi, 0.1)
F_exact = np.sin(x)
F_approx = cumtrapz(np.cos(x), x)

plt.figure(figsize = (10,6))
plt.plot(x, F_exact)
plt.plot(x[1::], F_approx)
plt.grid()
plt.tight_layout()
plt.title('$F(x) = \int_0^{x} cos(y) dy$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['Exact with Offset', 'Approx'])
plt.show()
```
