**Section 22.1(ODE Initial Value Problem Statement)**

A relationship between a function f(x), its independent variable, x, and any number of its derivative is a differentiual equation. A differential equation where the independent variables and the derivatives are in one dimension is called an ordinary differential equation or ODE.

d^(n)f(x)/dx^n

A general form differential equation is used to describe a partial differential equation which is where x is a vector that contains independent variables x1,x1, ... xm, and the partial derivatives can be of any order and with respect to any variable.

A function g(x) that satisfies the differential equation is called a general solution.

**Section 22.2(Reduction of Order)**

An important step to solve higher order differential equations is to reduce the order of the differential equation to first order. This is very helpful since most numerical methods for solving initial value problems were made specifically to solve first order differential equations.

Note that ODEs can be displayed such as:
f^(n)(t) = F(t,f(t),f^(1)(t),f^(2)(t), ..., f^(n-1)(t))

**Section 22.3(The Euler Method)**

If we let dS(t)/dt = F(t,S(t)) be an explicitly defined first order ODE and t be a numerical grid of the interval [t(0),t(f)] with spacing h then without loss of generality assume t(0)=0 and t(f)=Nf for some positive integer, N. With this we can compute an approximation for the state at S(t(j+1)) given the state at S(t(j)). This is formula is known as the Explicit Euler Formula.

S(t(j+1))=S(t(j))+hF(t(j),S(t(j))).

```python
#The differential equation df(t)/dt = e^(t) with the initial condition f(0)=-1 has the exact solution f(t)=e^(t)-2
#Using the Explicity Euler Formula approximate the solution to this intitial value problem between 0 and 1 in increments of 0.01.

import numpy as np
import matplotlib.pyplot as plt

# Define parameters
f = lambda t, s: np.exp(t) # ODE
h = 0.01 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = -1 # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'b--', label='Approximate')
plt.plot(t, np.exp(t)-2, 'g', label='Exact')
plt.title('Approximate and Exact Solution for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='upper left')
plt.show()
```
